## Creating a List

```python
# empty
print([])

# 1D --> Homogenous
print([1,2,3,4])

# 2D
print([1,2,3,[4,5]])

# 3D
print([[[1,2], [5,6]],[[3,4],[7,8]]])

# Hetrogeneous list
print([1,3,'Data', True, 2+4j])

# using type string
print(list('Lahore'))
```

## Accessing Items from a List

```python
# Indexing

# positive
L = [1,2,3,4,5]
print(L[3])

L1 = [[[1,2], [5,6]],[[3,4],[7,8]]]
print(L1[0][1][0])

# negative
print(L[-4])
print(L1[-1][-1][-2])
```

```python
# Slicing

# positive
L = [1,2,3,4,5]
print(L[0:4])
print(L[2::2])

#negative
L = [1,2,3,4,5]
print(L[-6:-1:])
print(L[-5::2])
print(L[::-1])
```

## Adding Items to a List

```python
# append
# -> single item, last

L = [1,2,3,4]
# L.append('Data')
# print(L)
# L.append(True)
# print(L)

L.append([5,6,7])
print(L)
```

## Deleting items from a List

```python
# pop
# default last, work with index

L = [1,2,3,4,5]
# L.pop(2)

L.pop()
print(L)
```

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L, reverse = True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(5)
# L.count(1)
```

```python
# index
L = [1,2,1,3,4,1,5]

L.index(5)
```

```python
# reverse
L = [2,1,5,7,0]
print(L)

# permanently changes the list
L.reverse()
print(L)
```

```python
# sort (vs sorted)       -->sort: permanently changes the string
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)

L.sort()
print(L)
```

```python
# copy -> shallow
L = [2,1,5,7,0]

print(L)
print(id(L))

L1 = L.copy()
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
# through loop

L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
# through list comprehension
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]

[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50

[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,2,-3,4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```

## Disadvantages of Python Lists

- Slow
- Risky usage
- eats up more memory

```python
# risku usage
a = [1,2,3]
b = a

print(a)
print(b)

a.append(4)
print(a)
print(b)

# lists are mutable
```

```python
# avoiding risk
a = [1,2,3]
b = a.copy()

print(a)
print(b)

a.append(4)
print(a)
print(b)

# lists are mutable
```

```python
# Create 2 lists from a given list where
# 1st list will contain all the odd numbers from the original list and
# the 2nd one will contain all the even numbers

L = [1,2,3,4,5,6]

L_odd = [i for i in L if i%2 != 0]
L_even = [j for j in L if j%2 == 0]

print(L_odd)
print(L_even)
```

```python
# Write a program to merge 2 list without using the + operator
L1 = [1,2,3,4]
L2 = [5,6,7,8]
L1.extend(L2)

print(L1)
```

```python
# Write a program to replace an item with a different item if found in the list
L = [1,2,3,4,5,3]
# replace 3 with 300

L = [300 if i == 3 else i for i in L]

print(L)
```

```python
# Write a program that can convert a 2D list to 1D list

list_2d = [[1, 2], [3, 4], [5, 6]]

list_1d = [item for sub_item in list_2d for item in sub_item]
print(list_1d)
```

```python
# Write a program to remove duplicate items from a list

L = [1,2,1,2,3,4,5,3,4]
unique_list = []

for i in L:
  if i not in unique_list:
    unique_list.append(i)

print('After removing duplicates:', unique_list)
```

```python
# Write a program to check if a list is in ascending order or not

L = [1, 3, 4, 5, 9]   # try: L = [1, 3, 2, 5, 9]

if L == sorted(L):
  print("List is in ascending order")
else:
  print("List is not in ascending order")
```
